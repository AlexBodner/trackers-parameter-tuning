### Detection reader function

```python
def get_yolox_detections(frame_id, det_list):
    dets_per_frame = [d for d in det_list if d.split(",")[0] == str(frame_id)]

    dets = []

    for i, line in enumerate(dets_per_frame):
        det = line.split(",") # "frame_no,x1,y1,x2,y2,det_conf"

        x1 = float(det[1])
        y1 = float(det[2])
        x2 = float(det[3])
        y2 = float(det[4])
        conf = float(det[5])

        det = [x1, y1, x2, y2, conf]
        dets.append(det)

    return dets
```

```python
import os
import shutil

gt_base_dir = "TrackEval/data/gt/dancetrack"

# Flatten ground truth files for 'val' split
val_split_dir = os.path.join(gt_base_dir, "val")
for seq_folder in sorted(os.listdir(val_split_dir)):
    full_seq_path = os.path.join(val_split_dir, seq_folder)
    if os.path.isdir(full_seq_path):
        source_gt_path = os.path.join(full_seq_path, "gt", "gt.txt")
        destination_gt_path = os.path.join(gt_base_dir, f"{seq_folder}.txt")
        if os.path.exists(source_gt_path):
            shutil.copy(source_gt_path, destination_gt_path)
            print(f"Copied {source_gt_path} to {destination_gt_path}")
        else:
            print(f"Warning: {source_gt_path} not found.")

print("Ground truth files flattening complete.")
```
